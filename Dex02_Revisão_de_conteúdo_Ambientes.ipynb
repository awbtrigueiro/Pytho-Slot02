```python
# Criando ma função de utilização

def score(data:pd.Series)->float:
  """
    Creates a score to measure the global distribution of the game.

    Inputs:
      - data (pd.Series): Series rith the columns NA_Sales	EU_Sales	JP_Sales	Other_Sales	Global_Sales
    Outputs:
      - score (float): Global score 
  """
  score = (data[0] / 3 + data[1] + data[2] + data[3] * 4) / data[4]
  return score


df_vg_csv['score'] = df_vg_csv[['NA_Sales',	'EU_Sales',	'JP_Sales',	'Other_Sales',	'Global_Sales']].apply(lambda x: score(x), axis=1)
df_vg_csv.head()
```

```python
df_vg_csv.sort_values('score', ascending=False).head()
```
